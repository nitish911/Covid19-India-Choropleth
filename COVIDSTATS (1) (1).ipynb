```python
k=dict(df)
f=[]
p=[]
y=dict()
for l in k["State"]:
    f.append(l)
for m in k["Positive"]:
    p.append(m)
for i in range(len(f)):
    y[f[i]]=p[i] 
state=[]
positive=[]
for a in y.keys():
    if a=="Odisha":
        state.append("Orissa")
        continue
    if a=="Uttarakhand":
        state.append("Uttaranchal")
        continue
    state.append(a)
for b in y.values():
    positive.append(b)
covid=dict()
covid["States"]=state
covid["Positives"]=positive
pdf=pd.DataFrame(covid)
pdf
```
